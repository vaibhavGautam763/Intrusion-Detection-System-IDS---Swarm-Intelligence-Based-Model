MODELLING

```python
# Prepare data for training
df = df.drop(['target'], axis=1)
Y = df[['Attack Type']]
X = df.drop(['Attack Type'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import time
```

```python
# Scale features
sc = MinMaxScaler()
X = sc.fit_transform(X)
```

```python
# Split test and train data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
# Convert labels to numerical values
Y_train = Y_train.replace({'normal': 0, 'dos': 1, 'u2r': 2, 'r2l': 3, 'probe': 4}).values
Y_test = Y_test.replace({'normal': 0, 'dos': 1, 'u2r': 2, 'r2l': 3, 'probe': 4}).values
```

```python
# 4. MODELING

# Build a simple neural network model
model = Sequential()
model.add(Dense(30, input_dim=X_train.shape[1], activation='relu', kernel_initializer='random_uniform'))
model.add(Dense(5, activation='softmax'))  # 5 classes: normal, dos, probe, r2l, u2r
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
start_time = time.time()
model.fit(X_train, Y_train, epochs=10, batch_size=64, verbose=1)
end_time = time.time()
print('Training time: ', end_time - start_time)
```

```python
start_time = time.time()
Y_test_pred = model.predict(X_test)
end_time = time.time()
print("Testing time: ", end_time - start_time)
```

```python
# Calculate accuracy
train_accuracy = accuracy_score(Y_train, model.predict(X_train).argmax(axis=1))
test_accuracy = accuracy_score(Y_test, Y_test_pred.argmax(axis=1))
print("Train accuracy:", train_accuracy)
print("Test accuracy:", test_accuracy)
```

```python
# 5. PERFORMANCE METRICS

# Performance metrics for training accuracy
names = ['ANN']
values = [train_accuracy * 100]
plt.figure(figsize=(15, 3))
plt.subplot(131)
plt.ylim(80, 102)
plt.bar(names, values)
plt.title('Training Accuracy')
plt.show()
```

```python
# Performance metrics for testing accuracy
names = ['ANN']
values = [test_accuracy * 100]
plt.figure(figsize=(15, 3))
plt.subplot(131)
plt.ylim(80, 102)
plt.bar(names, values)
plt.title('Testing Accuracy')
plt.show()
```

```python
# Continuing from your previous code

# 6. TRANSFER LEARNING IMPLEMENTATION

# For the purpose of transfer learning, we will pre-train a model on a subset of the data,
# then use the pre-trained model weights to initialize a new model, freeze some layers,
# and fine-tune it on the full dataset.

# Step 1: Pre-train on a subset of the data
# Let's take 30% of the training data for pre-training
X_pretrain, X_remaining, Y_pretrain, Y_remaining = train_test_split(X_train, Y_train, test_size=0.7, random_state=42)

# Build the pre-training model
pretrain_model = Sequential()
pretrain_model.add(Dense(30, input_dim=X_pretrain.shape[1], activation='relu', kernel_initializer='random_uniform'))
pretrain_model.add(Dense(15, activation='relu', kernel_initializer='random_uniform'))
pretrain_model.add(Dense(5, activation='softmax'))  # 5 classes
pretrain_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Pre-train the model
start_time = time.time()
pretrain_model.fit(X_pretrain, Y_pretrain, epochs=10, batch_size=64, verbose=1)
end_time = time.time()
print('Pre-training time: ', end_time - start_time)

# Step 2: Transfer the learned weights to a new model
# Build a new model with the same architecture
transfer_model = Sequential()
transfer_model.add(Dense(30, input_dim=X_train.shape[1], activation='relu', kernel_initializer='random_uniform'))
transfer_model.add(Dense(15, activation='relu', kernel_initializer='random_uniform'))
transfer_model.add(Dense(5, activation='softmax'))

# Copy the weights from the pre-trained model to the new model
transfer_model.set_weights(pretrain_model.get_weights())

# Freeze the first layer to keep the learned features
transfer_model.layers[0].trainable = False

# Compile the transfer model
transfer_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 3: Fine-tune the model on the full training data
start_time = time.time()
transfer_model.fit(X_train, Y_train, epochs=10, batch_size=64, verbose=1)
end_time = time.time()
print('Fine-tuning time: ', end_time - start_time)

# Evaluate the transfer model
start_time = time.time()
Y_test_pred = transfer_model.predict(X_test)
end_time = time.time()
print("Testing time after transfer learning: ", end_time - start_time)

# Calculate accuracy
train_accuracy = accuracy_score(Y_train, transfer_model.predict(X_train).argmax(axis=1))
test_accuracy = accuracy_score(Y_test, Y_test_pred.argmax(axis=1))

print("Train accuracy after transfer learning:", train_accuracy)
print("Test accuracy after transfer learning:", test_accuracy)

# 7. PERFORMANCE METRICS AFTER TRANSFER LEARNING

# Performance metrics for training accuracy
names = ['ANN with Transfer Learning']
values = [train_accuracy * 100]
plt.figure(figsize=(15, 3))
plt.subplot(131)
plt.ylim(80, 102)
plt.bar(names, values)
plt.title('Training Accuracy after Transfer Learning')
plt.show()

# Performance metrics for testing accuracy
names = ['ANN with Transfer Learning']
values = [test_accuracy * 100]
plt.figure(figsize=(15, 3))
plt.subplot(131)
plt.ylim(80, 102)
plt.bar(names, values)
plt.title('Testing Accuracy after Transfer Learning')
plt.show()
```

```python
from sklearn.metrics import classification_report

# Modify the model to return history for plotting learning curves
history = transfer_model.fit(X_train, Y_train, epochs=10, batch_size=64, verbose=1, validation_data=(X_test, Y_test))

# 1. AUC Score Calculation
# Generate predicted probabilities for the test set
Y_test_prob = transfer_model.predict(X_test)

# Since AUC requires the probabilities, we calculate it for each class.
# You can calculate the average AUC across all classes
auc_scores = []
for i in range(5):  # 5 classes: normal, dos, u2r, r2l, probe
    auc = roc_auc_score((Y_test == i).astype(int), Y_test_prob[:, i])
    auc_scores.append(auc)

# Calculate average AUC score
avg_auc_score = np.mean(auc_scores)
print(f"Average AUC Score: {avg_auc_score}")

# 2. Plot ROC Curves for each class
plt.figure(figsize=(10, 8))
for i in range(5):
    fpr, tpr, _ = roc_curve((Y_test == i).astype(int), Y_test_prob[:, i])
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {auc_scores[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Each Class')
plt.legend(loc='lower right')
plt.show()

# 3. Learning Curves (Accuracy and Loss)

# Plotting training and validation accuracy
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plotting training and validation loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 4. Classification Report
# Generate class predictions (not probabilities)
Y_test_pred = np.argmax(Y_test_prob, axis=1)

# Print classification report
print("Classification Report:\n")
print(classification_report(Y_test, Y_test_pred))
```
